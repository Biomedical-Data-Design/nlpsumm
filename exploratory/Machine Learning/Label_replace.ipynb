```python
import numpy as np
import pandas as pd
```

```python
#read txt file
def read_txt(filename):
    with open(filename, 'r') as f:
        data = f.readlines()
    return data

traininglist = read_txt('E:/JHU/课程/datadesign/NLP/machine_learning/trainingdata.txt')
#print(list)
for i in range(len(traininglist)):
    traininglist[i] = traininglist[i].split()
#print(list)
trainingdataframe = pd.DataFrame(traininglist, columns=['text', 'label'])
print(trainingdataframe.shape)
trainingdataframe.head()
```

```python
df_training = trainingdataframe.fillna(method='ffill')
df_training.text.nunique(), df_training.label.nunique()
```

```python
df_training.groupby('label').size().reset_index(name='counts')
```

```python
#将B-PHI和I-PHI合并为O
df_training['label'] = df_training['label'].replace(['B-PHI', 'I-PHI'], 'O')
df_training.groupby('label').size().reset_index(name='counts')
```

```python
#存储为txt文件
df_training.to_csv('E:/JHU/课程/datadesign/NLP/machine_learning/trainingdata_noPHI.txt', sep='\t', index=False, header=False)
```
